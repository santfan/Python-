# Чат-бот с ИИ моделью falcon-7b-instruct
Модели генерации текста типа GPT-3 предназначены для генерации текстов.<br> Генерация тестов в таких моделях происходит следющим образом:

*   Следует понимать что модели работают с числовыми значениями. Поэтому вводимый текст должен быть преобразован в некоторую последавательность чисел.
*   Для прееоразования текста в число происходит используют словари.
*   Словари формируются из обработаных текстов.
*   Обратока текстов сводится к :

  1.   Лемматизации текста
  2.   Токенизации текста
---
Прежде чем писать о лемматизации и токенизации следует понимать что любой текст содержит много предлогов, союзов, местоимений и т.п.. Все эти элементы конечно же нужны, но смысловой нагрузки они не несут. Поэтому текст или коллекцию текстов необходимо избавить от таких элементов текста. Все это можно сделать обычным циклом
```python
# Cписок элементов котрые нужно удалить
repl_elem = ['в', 'и', 'из', 'к', 'на', 'что', 'как', 'я', 'ты', 'они',.... и т.д и т.п.]
# Открываем файл
with open('file_name.txt', 'r') as file
# Пробегаем окрытый файл
for elm in file:
  if elem in repl_elem:
    replace(elem)
# Понятно что этот код не отработает и приведен только для иллюстрации самой идеи
```

После того как из текста удалили лишнюю (с точки зрения смысловой нагрузки) можно приступать к лемматизации текста
## Лемматизация текста.
Идею **лемматизации** текста сформулировал математик Лемм. Идея заключается в следующем: <br>
Если посмотреть любой текст, то выянится, что используются различные формы (к примеру) глаголов, которые по сути обозначают одно и тоже. Галаголы "сделал", "сделала", "делают", "делаем" это собственнно с точки зрения смысла одно и тоже. Это наводит на мысль что можно заменить все эти глаголы одним "делать". Такая замена сильно сократит размер словаря не изменяя при этом смыловой функциональности.<br>
Лемматизированный словарь можно подвергнуть другой не менее важной, а скорее всего самой важной процедуре - процедуре **токенизации**  
## Токенизация текста.
Итак мы уже говорили что языковые модели точно так же как и другие типы моделей.Работают с числами. Так вот процесс токенизации это собственно процесс присвоения каждому слову лемматизированого словаря некотрого уникального числового значения. Собственного само число присваивается исходя из частотного анализа текста. Т.е. чем чаще встречается слово тем ниже его числовое значение. Но тут надо понимать что термины, которые и определяют смысл теста с точки зрения профессионального контента, встречаются не так часто как те же самые глаголы, но несут бОльшую смысловую нагрузку. Исходя из этого понимания, следует, что токены одной модели отличаются от токенов другой. Иначе говоря,токены словаря физика обзначают слова, а токены словаря химика имеющие теже числовые значения ссылаются на другие.

После того как мы рассмотрели приципы **лемматизации** и **токенизации** можно переходить к тому что мы собираемся получить в итоге. Итак LLM модели типа GPT-3 это большие языковые модели собственно говоря термин LLM (long laguage model) говорит сам за себя. Такие модели требуют огромных ресурсов для их создания. Но и возможности их так же велики. Часто в дополнение к LLM используют так называемые модели с инструкцией. Модели LLM могут по одному предложению, определив контекст запроса, могут выдать не то что абзац, но целую лекцию. Модели с инструкцией вступают в диалог с пользоватем. Грубо говоря модели с инструкцией можно рассматривать как личного промт-инженера. Такие модели обучаются на собственных словарях - поэтому они специализированы. Пользователь задает вопрос модели с инструкией. Модель токенизирует запрос пересылает его LLM модели. И, получив ответ, преобразует получене токены в текст выдает его пользователю.

## Итоговый код

````python
from transformers import AutoTokenizer, pipeline
import torch

model = "tiiuae/falcon-7b-instruct"
tokenizer = AutoTokenizer.from_pretrained(model)
pipeline = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    torch_dtype=torch.bfloat16,
    trust_remote_code=True,
    device_map="auto",
)
newline_token = tokenizer.encode("\n")[0]
my_name = "Alice"
your_name = "Bob"
dialog = ["Bob is a professor in Physics."]

while True:
    user_input = input("> ")
    dialog.append(f"{my_name}: {user_input}")
    prompt = "\n".join(dialog) + f"\n{your_name}: "
    sequences = pipeline(
        prompt,
        truncation=True,
        max_length=500,
        do_sample=True,
        top_k=10,
        num_return_sequences=1,
        return_full_text=False,
        eos_token_id=newline_token,
        pad_token_id=tokenizer.eos_token_id,
    )
    print(sequences[0]['generated_text'])
    dialog.append("Bob: "+sequences[0]['generated_text'])
````
